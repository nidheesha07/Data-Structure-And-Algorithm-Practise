Lambda Functions

```python
a = 'GeeksforGeeks'
upper = lambda x: x.upper()  
print(upper(a))
```

Condition Checking

```python
check = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(check(1))   
print(check(-1))  
print(check(0))
```

List Comprehension

```python
func = [lambda arg=x: arg * 10 for x in range(1, 5)]
for i in func:
    print(i())
```

Returning Multiple Results

```python
calc = lambda x, y: (x + y, x * y)
res = calc(3, 4)
print(res)
```

 filter()

```python
c = [1, 2, 3, 4, 5, 6]
even = filter(lambda x: x % 2 == 0, c)
print(list(even))
```

map()

```python
a = [1, 2, 3, 4]
double = map(lambda x: x * 2, a)
print(list(double))
```

reduce()

```python
from functools import reduce
a = [1, 2, 3, 4]
mul = reduce(lambda x, y: x * y, a)
print(mul)
```

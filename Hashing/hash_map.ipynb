```python
student = {
    "Name": "John",
    "Age": 20,
    "Marks": 95
}

print(student)
```

```python
student = {}

# Insert
student["Name"] = "John"
student["Age"] = 20
student["Marks"] = 95

print(student)

# Access
print(student["Name"])

# Update
student["Marks"] = 98

print(student)

# Delete
del student["Age"]

print(student)
```

Traversing a List

```python
nums = [10, 20, 30, 40, 50]
for num in nums:
    print(num)
```

2. Finding Maximum Element

```python
nums = [4, 8, 2, 15, 7]
max = nums[0]
for num in nums:
    if num > max:
        max = num
print(max)
```

3. Finding Minimum Element

```python
nums = [4, 8, 2, 15, 7]
minimum = nums[0]
for num in nums:
    if num < minimum:
        minimum = num
print(minimum)
```

4. Sum of Elements

```python
nums = [1, 2, 3, 4, 5]
total = 0
for num in nums:
    total += num
print(total)
```

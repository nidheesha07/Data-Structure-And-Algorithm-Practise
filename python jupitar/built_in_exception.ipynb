1. BaseException

```python
try:
    num = int("abc")
except BaseException as e:
    print("Error:", e)
    print("Error:", type(e).__name__)
```

```python
try:
    a=10
    result = a / 0
except BaseException as e:
    print("Error:", type(e).__name__)
    print("Error:", e)
```

```python
try:
    num = int(10)
    print(num)
except BaseException as e:
    print("Error:", e)
```

2. Exception

```python
try:
    amount = 7000

    if amount > 5000:
        raise Exception("Insufficient Balance")

    print("Withdrawal Successful")

except Exception as e:
    print("Error:", e)
```

3. Arithmetic Error

```python
try:
    a=10
    b=0
    result=a/b
    print(result)

except ArithmeticError as e:
    print("Arithmetic Error Occurred:", e)
```

4. ZeroDivisionError

```python
try:
    a=10
    b=0
    result=a/b
    print(result)

except ZeroDivisionError as e:
    print("Zero Division Error Occurred:", e)
```

5. OverflowError

```python
import math
try:
    result = math.exp(1000)  # e^1000
    print(result)

except OverflowError as e:
    print("Overflow Error:", e)
```

6. FloatingPointError

```python
try:
    raise FloatingPointError("Floating point calculation failed")
except FloatingPointError as e:
    print("Error:", e)
```

7.AssertionError

```python
age = 20
assert age >= 18
print("Eligible to vote")
```

```python
age = 15
assert age >= 18
print("Eligible to vote")
```

8. AttributeError

```python
name = "Python"
print(name.append("!"))
```

```python
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Rahul")

print(s.age)
```

9. IndexError

```python
my_list = [1, 2, 3]

try:
    element = my_list[5]
except IndexError as e:
    print(e)
```

10. KeyError

```python
d = {"key1": "value1"}

try:
    val = d["key2"]
except KeyError as e:
    print(e)
```

11. MemoryError

```python
try:
    a = [1] * (10**10)
except MemoryError as e:
    print(e)
```

12. NameError

```python
try:
    print(var)
except NameError as e:
    print(e)
```

13. OSError

```python
try:
    file = open("data.txt", "r")
except OSError as e:
    print("Error:", e)
```

1. BaseException

```python
try:
    num = int("abc")
except BaseException as e:
    print("Error:", e)
    print("Error:", type(e).__name__)
```

```python
try:
    a=10
    result = a / 0
except BaseException as e:
    print("Error:", type(e).__name__)
    print("Error:", e)
```

```python
try:
    num = int(10)
    print(num)
except BaseException as e:
    print("Error:", e)
```

2. Exception

```python
try:
    amount = 7000

    if amount > 5000:
        raise Exception("Insufficient Balance")

    print("Withdrawal Successful")

except Exception as e:
    print("Error:", e)
```

3. Arithmetic Error

```python
try:
    a=10
    b=0
    result=a/b
    print(result)

except ArithmeticError as e:
    print("Arithmetic Error Occurred:", e)
```

4. ZeroDivisionError

```python
try:
    a=10
    b=0
    result=a/b
    print(result)

except ZeroDivisionError as e:
    print("Zero Division Error Occurred:", e)
```

5. OverflowError

```python
import math
try:
    result = math.exp(1000)  # e^1000
    print(result)

except OverflowError as e:
    print("Overflow Error:", e)
```

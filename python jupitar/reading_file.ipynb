Reading a file

```python
file = open("Helloworld.txt", "r")
content = file.read()
print(content)
file.close()
```

Using with statement

```python
with open("Helloworld.txt","r") as file:
    content=file.read()
    print(content)
```

Reading a File Line by Line

```python
with open("Helloworld.txt","r") as file:
    for line in file:
        print(line)
```

```python
with open("Helloworld.txt","r") as file:
    for line in file:
        print(line.strip()) #it removes extra newline characters
```

```python
with open("Helloworld.txt","r") as file:
    print(file.readline())
    print(file.readline())
    print(file.readline())
```

Reading Binary Files

```python
#This is used to create binary file
file = open("example.bin", "w")
file.write("This is binary file")
file.close()
```

```python
#reads the file
file = open("example.bin", "rb") #rb is binary read mode
content = file.read()
print(content)
file.close() # b in output is data is a bytes object
```

Reading Specific Parts of a File

```python
file = open("Helloworld.txt", "r")
content = file.read(10)
print(content)
file.close()
```

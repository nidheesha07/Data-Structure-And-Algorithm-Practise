map()

```python
a = [1, 2, 3, 4]
b = list(map(lambda x: x + 2, a))
print(b)
```

```python
a = [1, 2, 3, 4]
b = tuple(map(lambda x: x + 2, a))
print(b)
```

filter()

```python
a=[1,2,3,4,5,6]
b=filter( lambda x:x%2!=0 , a)
print(list(b))
```

```python
a=["Hello ", "world"]
b=filter( lambda x:len(x)>5, a)
print(list(b))
```

```python
from functools import reduce
a=["Hello " , "world"]
b=reduce(lambda x,y: x+y , a)
print(b)
```

inner function()

```python
def func1():
    msg="Hello world"
    def func2():
        print(msg)
    func2()
func1()
    
```

```python
def calc():
    def add(a,b):
        print(a+b)
    def sub(a,b):
        print(a-b)
    add(10,5)
    sub(10,5)
calc()
```

decorators

```python
def decorator(func):
    def wrapper():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrapper

@decorator
def func2():
    print("Welcome")
def greet():
    print("Hello World")
func2()
greet()
```

```python
def decorator(func):
    def wrapper():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrapper

@decorator
def greet():
    print("Hello World")
def func2():
    print("Welcome")
func2()
greet()
```

Polymorphism

1. Compile-time Polymorphism

```python
class Calculator:
    def multiply(self, a=1, b=1, *args):
        result = a * b
        for num in args:
            result *= num
        return result
calc = Calculator()
print(calc.multiply())            
print(calc.multiply(4))           
print(calc.multiply(2, 3))       
print(calc.multiply(2, 3, 4))
```

Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def info(self):
        print("Animal name:", self.name)

class Dog(Animal):
    def sound(self):
        print(self.name, "barks")

d = Dog("Buddy")
# Inherited method
d.info()     
d.sound()
```

```python
#super() is used to call constructor of parent clas frm child cls
class Student:
    def __init__(self, name):
        self.name = name
        print("Student Constructor Called")
class EngineeringStudent(Student):
    def __init__(self, name):
        super().__init__(name)
        print("Engineering Student Constructor Called")
s1 = EngineeringStudent("Nidheesha")
```

Abstract Base Class

```python
from abc import ABC, abstractmethod
class Student(ABC):
    @abstractmethod
    def calculate_grade(self):
        pass
class EngineeringStudent(Student):
    def calculate_grade(self):
        print("Engineering Grade: A")
s1 = EngineeringStudent()
s1.calculate_grade()
```

```python
from abc import ABC, abstractmethod
class Student(ABC):
    @abstractmethod
    def study(self):
        pass
class EngineeringStudent(Student):
    def study(self):
        print("Studying Data Structures")
s1 = EngineeringStudent()
s1.study()
```

Encapsulation

```python
class Student:
    def __init__(self, name, marks):
        self.__name = name      #private variable (__name)
        self.__marks = marks
    def get_marks(self):    #Used to read the private variable.
        return self.__marks
    def set_marks(self, marks):   #Used to modify the private variable safely.
        self.__marks = marks
s1 = Student("Nidheesha", 90)
#print(s1.__marks) - error coz __marks is private
print(s1.get_marks())
s1.set_marks(95)
print(s1.get_marks())
```

public , private , protected

```python
class Student:

    def __init__(self):
        self.name = "Nidheesha"      # Public #can be accessed frm anywhere
        self._course = "BCA"         # Protected  #can be used within class
        self.__marks = 95            # Private  #cant access frm outside cls


s1 = Student()

print(s1.name)
print(s1._course)
# print(s1.__marks)   #  attribute Error
```

public member

```python
class Student:
    def __init__(self):
        self.name = "Nidheesha"   # Public variable
s1 = Student()
print(s1.name)
```

protected member

```python
class Student:
    def __init__(self):
        self._course = "BCA"   # Protected variable
s1 = Student()
print(s1._course)
```

Private member

```python
class Student:
    def __init__(self):
        self.__marks = 95   # Private variable
s1 = Student()
print(s1.__marks)
```

Accessing Private Data Using a Method

```python
class Student:
    def __init__(self):
        self.name = "Nidheesha"
        self._course = "BCA"
        self.__marks = 95
    def get_marks(self):
        return self.__marks
s1 = Student()
print(s1.get_marks())
```

```python
class Student:
    def public_method(self):
        print("Public Method")
    def _protected_method(self):
        print("Protected Method")
    def __private_method(self):
        print("Private Method")
    def display(self):
        self.public_method()
        self._protected_method()
        self.__private_method()
s1 = Student()
s1.display()
```

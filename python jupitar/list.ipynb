Using Square Brackets:

```python
a = [1, 2, 3]
print(a)
b = ["apple", "banana"]
print(b)
```

Using list() Constructor

```python
a = list((1, 2, 3, 'apple', 3.14))  
print(a)
b = list("hello world")
print(b)
```

Creating List with Repeated Elements:

```python
print([1]*2)
print([7]*3)
```

Internal Representation of Lists

```python
a = [1, 2, "hello"]
print(a[1])   # index-based
print(a)
```

Accessing List Elements

```python
a = [1 ,2 , 3]
print(a[0])
print(a[-1])
```

Adding Elements into List

```python
a = [1, 2]
a.append(3)
print(a)
```

```python
a = [1, 3]
a.insert(1, 2)
print(a)
```

```python
a = [1, 2]
a.extend([3, 4])
print(a)
```

Updating Elements into List

```python
a = [10, 15, 30, 40, 50]
a[1] = 20 
print(a)
```

Removing Elements from List

```python
a = [10, 20, 30]
a.remove(10)
print(a)
```

```python
a = [1, 2, 3]
a.pop()
print(a)
```

```python
a = [1, 2, 3]
a.pop(1)
print(a)
```

```python
a = [1, 2, 3]
del a[0]
print(a)
```

```python
a = [1, 2, 3]
del a
print(a)
```

```python
a = [1, 2, 3]
a.clear()
print(a)
```

Iterating Over Lists

```python
a = [10 , 20, 30]
for i in a:
    print(i)
```

Nested Lists

```python
a = [[1, 2], [3, 4]]
print(a)
print(a[0])
print(a[1][0])
```

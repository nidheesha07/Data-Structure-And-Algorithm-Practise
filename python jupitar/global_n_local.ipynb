Local Variables

```python
def greet():
    msg = "Hello world!"
    print(msg)

greet()
```

```python
def greet():
    msg = "Hello world!"
    print("Inside the function:",msg)

greet()
print("Outside the function: ",msg)
```

Global Variables

```python
msg = "Hello world!"
def greet():
    print("Inside the function:",msg)

greet()
print("Outside the function: ",msg)
```

Use of Local and Global variables

```python
msg = "Hello world!"
def greet():
    msg = "This is python programming"
    print("Inside the function:",msg)

greet()
print("Outside the function: ",msg)
```

Modifying Global Variables Inside a Function

Without global

```python
msg = "Hello world!"
def greet():
    msg += "This is python programming"
    print("Inside the function:",msg)

greet()
print("Outside the function: ",msg)
```

with global

```python
msg = "Hello world!"
def greet():
    global msg
    msg += "This is python programming"
    print("Inside the function:",msg)

greet()
print("Outside the function: ",msg)
```

```python
msg = "Hello world!"
def greet():
    global msg
    msg += "This is python programming"
    print("Inside the function:",msg)
print("Outside the function: ",msg)
greet()
```

Global vs Local with Same NameGlobal vs Local with Same Name

```python
a = 1  # Global variable

def f():
    print("f():", a)  # Uses global a

def g():
    a = 2  # Local shadows global
    print("g():", a)

def h():
    global a
    a = 3  # Modifies global a
    print("h():", a)

print("global:", a)
f()
print("global:", a)
g()
print("global:", a)
h()
print("global:", a)
```

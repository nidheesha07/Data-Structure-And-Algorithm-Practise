```python
#Changing value does not affect order
from collections import OrderedDict
od = OrderedDict([('a', 1), ('b', 2), ('c', 3), ('d', 4)])
od['c'] = 5  
print(od)
print(list(od.items()))
for k, v in od.items():
    print(k, v)
```

```python
#Equality checks consider order
from collections import OrderedDict
od1 = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
od2 = OrderedDict([('c', 3), ('b', 2), ('a', 1)])
print(od1 == od2)
```

```python
#Reversing an OrderedDict
from collections import OrderedDict
od = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
rev_od=list((reversed(od.items())))
print(rev_od)
```

Defaultdict 

```python
from collections import defaultdict

words = ["apple", "banana", "apple"]

count = defaultdict(int)

for word in words:
    count[word] += 1

print(dict(count))
```

```python
# Default Factory
from collections import defaultdict
d = defaultdict(list)

d['fruits'].append('apple')
d['vegetables'].append('carrot')
print(list(d.items()))
```

```python
#Using int Default Factory
from collections import defaultdict
d = defaultdict(int)
print(d['juices'])
```

```python
#Using List as Default Factory
from collections import defaultdict
d = defaultdict(list)
print(d['juices'])
```

```python
#Using str default factory
from collections import defaultdict
d = defaultdict(str)
print(d['juices'])
```

```python
from collections import defaultdict
d = defaultdict(lambda: "Not Present")
d["a"] = 1
d["b"] = 2
print(d.__missing__('x'))
print(d.__missing__('d'))
print(d['a'])
```

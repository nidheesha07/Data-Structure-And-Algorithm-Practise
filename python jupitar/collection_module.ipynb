Heapq

```python
import heapq
numbers = [40, 10, 30, 20, 50]
heapq.heapify(numbers)
print(numbers)
```

```python
import heapq
fruits = ["mango" , "apple" , "orange" , "banana"]
heapq.heapify(fruits)
print(fruits)
```

```python
import heapq
numbers = [40, 10, 30, 20, 50]
heapq.heapify(numbers)
heapq.heappush(numbers, 25)
print(numbers)
```

```python
import heapq
numbers = [40, 10, 30, 20, 50]
heapq.heapify(numbers)
smallest=heapq.heappop(numbers)
print(smallest)
print(numbers)
```

```python
import heapq
numbers = [10, 50, 20, 80, 30, 70]
largest = heapq.nlargest(4, numbers)
print(largest)
```

```python
import heapq
numbers = [10, 50, 20, 80, 30, 70]
smallest = heapq.nsmallest(3, numbers)
print(smallest)
```

Deque

```python
from collections import deque
number=deque([10,20,30,40,50])
number.append(5) #adds element at the end of the list
print(number)
```

```python
from collections import deque
dq = deque([10, 20, 30])
dq.append(40)
print(dq)
```

```python
from collections import deque
number=deque([10,20,30,40,50])
number.appendleft(5) #adds element at the beginnning of the list
print(number)
```

```python
from collections import deque
dq = deque([10, 20, 30])
dq.appendleft(40)
print(dq)
```

```python
from collections import deque
number=deque([10,20,30,40,50])
number.pop()
print(number)
```

```python
from collections import deque
number=deque([10,20,30,40,50])
number.popleft()
print(number)
```

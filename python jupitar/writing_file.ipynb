Overwrite an existing file

```python
with open("Helloworld.txt" , "w") as file:
    file.write("Data structure and algorithm")
```

Append to an existing file

```python
with open("Helloworld.txt" , "a") as file:
    file.write("\nHello World")
```

Create a File Only If It Does Not Exist

```python
# x creates a new file
#raises error if file already exists
with open("student.txt", "x") as file:
    file.write("Nidheesha")
```

```python
with open("student.txt", "w") as file:
    file.write("Nidheesha\n")
    file.write("Python\n")
    file.write("DSA\n")
```

```python
lines = ["Hello\n", "world\n", "...\n"]

with open("student.txt", "w") as file:
    file.writelines(lines)
```

Writing to a Binary File

```python
data= b"Hello world"
with open("example.bin" , "wb") as file:
    file.write(data)
```

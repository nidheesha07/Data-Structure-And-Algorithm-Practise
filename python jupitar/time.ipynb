```python
arr = [10, 20, 30, 40, 50, 60, 70]

target = 50

low = 0
high = len(arr) - 1

while low <= high:
    mid = (low + high) // 2

    if arr[mid] == target:
        print("Found")
        break

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1
```

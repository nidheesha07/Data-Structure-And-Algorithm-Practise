Creating a Tuple

```python
a=()
b=("Hello" , "World")
c=(1 , 2, 3)
d=("Hello")
print(a)
print(b)
print(c)
print(d)
```

Creating a Tuple with Mixed Datatypes

```python
a= (7, "nidhi" , 3.14 , [1,2,3] , {'key':'value'})
print(a)
```

Accessing of Tuples

```python
a= "Hello world"
b= tuple("Hello")
print(a[0])
print(b[3])
n = "Hello" , "World"
x , y = n
print(x)
print(y)
```

Concatenation of Tuples

```python
a= ("Hello" , "world")
b=(1,2,3)
print(a+b)
```

Slicing of Tuple

```python
a="Hello world"
print(a[2:])
print(a[:-2])
```

Deleting a Tuple

```python
a="Hello world"
del a
print(a)
```

Tuple Unpacking with Asterisk (*)

```python
a= (1,2,3,4,5)
x, *y, z=a
print(x)
print(y)
print(z)
```

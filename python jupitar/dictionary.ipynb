Creating a Dictionary

```python
a = {"x": 1, "y": 2}
print(a)

b = dict(name="Sam", age=20)
print(b)
```

Adding and Updating Dictionary Items

```python
a={"name":"sam"}
a["name"] = "Nidhi" #updating
a["age"]=19 #adding
print(a)
```

Removing Dictionary Items

```python
d = {"a": 1, "b": 2}
del d["a"]
print(d)
```

```python
a = {"a": 1, "b": 2}
b = a.pop("a")
print(a)
print(b)
```

```python
a = {"a": 1, "b": 2}
print(a.popitem())
```

```python
d = {"a": 1, "b": 2}
d.clear()
print(d)
```

Iterating Through a Dictionary

```python
#iterate keys
d = {"a": 1, "b": 2}
for key in d:
    print(key)
```

```python
#iterating keys n values
d = {"a": 1, "b": 2}
for key, value in d.items():
    print(key, value)
```

For Loop

```python
n = 4
for i in range(0, n):
    print(i)
```

Iterating by Index of Sequences

```python
a = ["geeks", "for", "geeks"]
for i in range(len(a)):
    print(a[i])
```

While Loop

```python
c = 0
while (c < 3):
    c = c + 1
    print("Hello Geek")
```

Nested Loops

```python
for i in range(1, 5):
    for j in range(i):
        print(i, end=' ')
    print()
```

Calling a Function

```python
def func():
    print("Welcome to python programming")
func()
```

Function Arguments

```python
def add(a,b):
    sum=a+b
    return sum
result=add(10,20)
print("Result = ",result)
```

Default argument

```python
def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10,20)
```

```python
def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)
```

 Keyword Arguments

```python
def student(fname, lname):
    print(fname, lname)

student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')
```

Positional Arguments

```python
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Olivia", 27)

print("Case-2:")
nameAge(27, "Olivia")
```

Function within Functions

```python
def f1():
    print("Hello world")
    def f2():
        print("welcome to python programming")
    f2()
f1()
```

Return Statement

```python
def add(a,b):
    sum=a+b
    return sum
result=add(10,20)
print("Result = ",result)
```

Pass by Value

```python
def modify(x):
    x = x + 10
    print("Inside Function:", x)

num = 20
modify(num)

print("Outside Function:", num)
```

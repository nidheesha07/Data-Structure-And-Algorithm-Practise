```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

compile time polymorphism

```python
class Calculator:
    def multiply(self, a=1, b=1, *args):
        result = a * b
        for num in args:
            result *= num
        return result
calc = Calculator()
print(calc.multiply())            
print(calc.multiply(4))           
print(calc.multiply(2, 3))       
print(calc.multiply(2, 3, 4))
```

Run time polymorphism

```python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

dog = Dog()
dog.sound()
```

```python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def noise(self):
        print("Bark")

dog = Dog()
dog.sound()
```

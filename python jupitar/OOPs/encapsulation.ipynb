Encapsulation

```python
class Student:
    def __init__(self, name, marks):
        self.__name = name      #private variable (__name)
        self.__marks = marks
    def get_marks(self):    #Used to read the private variable.
        return self.__marks
    def set_marks(self, marks):   #Used to modify the private variable safely.
        self.__marks = marks
s1 = Student("Nidheesha", 90)
#print(s1.__marks) - error coz __marks is private
print(s1.get_marks())
s1.set_marks(95)
print(s1.get_marks())
```

public , private , protected

```python
class Student:

    def __init__(self):
        self.name = "Nidheesha"      # Public #can be accessed frm anywhere
        self._course = "BCA"         # Protected  #can be used within class
        self.__marks = 95            # Private  #cant access frm outside cls


s1 = Student()

print(s1.name)
print(s1._course)
# print(s1.__marks)   #  attribute Error
```

public member

```python
class Student:
    def __init__(self):
        self.name = "Nidheesha"   # Public variable
s1 = Student()
print(s1.name)
```

protected member

```python
class Student:
    def __init__(self):
        self._course = "BCA"   # Protected variable
s1 = Student()
print(s1._course)
```

Private member

```python
class Student:
    def __init__(self):
        self.__marks = 95   # Private variable
s1 = Student()
print(s1.__marks)
```

Accessing Private Data Using a Method

```python
class Student:
    def __init__(self):
        self.name = "Nidheesha"
        self._course = "BCA"
        self.__marks = 95
    def get_marks(self):
        return self.__marks
s1 = Student()
print(s1.get_marks())
```

```python
class Student:
    def public_method(self):
        print("Public Method")
    def _protected_method(self):
        print("Protected Method")
    def __private_method(self):
        print("Private Method")
    def display(self):
        self.public_method()
        self._protected_method()
        self.__private_method()
s1 = Student()
s1.display()
```

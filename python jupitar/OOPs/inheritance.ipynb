Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def info(self):
        print("Animal name:", self.name)

class Dog(Animal):
    def sound(self):
        print(self.name, "barks")

d = Dog("Buddy")
# Inherited method
d.info()     
d.sound()
```

```python
#super() is used to call constructor of parent clas frm child cls
class Student:
    def __init__(self, name):
        self.name = name
        print("Student Constructor Called")
class EngineeringStudent(Student):
    def __init__(self, name):
        super().__init__(name)
        print("Engineering Student Constructor Called")
s1 = EngineeringStudent("Nidheesha")
```

Example 2

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print("Brand:", self.brand)

class Car(Vehicle):
    def start(self):
        print(self.brand, "car started")

c = Car("Toyota")

c.show_brand()
c.start()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print("Vehicle Constructor Called")

class Car(Vehicle):
    def __init__(self, brand):   #constructor of a child class
        super().__init__(brand)
        print("Car Constructor Called")

c = Car("Toyota")
```

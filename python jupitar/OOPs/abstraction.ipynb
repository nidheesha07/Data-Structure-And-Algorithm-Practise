```python
class BankAccount:
    def __init__(self):
        self.__balance = 0  # Private variable (hidden detail) [2]

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}")
        else:
            print("Invalid transaction")

# User only sees deposit and withdraw, not the internal balance logic [2, 3]
account = BankAccount()
account.deposit(500)
```

```python
f = open("Helloworld.txt", "w")
f.write("This is python programming")
f.close()
```

File Handling

```python
f=open("Helloworld.txt","r") #opening a file
print(f.read())
f.close() #closing a file
```

```python
f=open("Helloworld.txt","r")
print("Filename :",f.name)
print("Mode : ",f.mode)
print("Is closed? ",f.closed)
f.close()
print("Is closed? ",f.closed)
```

Reading a File

```python
#THis reads the content present in the file
file = open("Helloworld.txt", "r")
content = file.read()
print(content)
file.close()
```

Writing a File

```python
#This creates n overwrites content to the file
file = open("Helloworld.txt", "w") #write mode
file.write("\nHello Wrorld!!\n")
file.write("Welcome to Python Programming")
file.close()
```

```python
#This write content to the exsisting file
file = open("Helloworld.txt", "a") #append mode
file.write("\nHello Wrorld!!\n")
file.write("Welcome to Python Programming")
file.close()
```

Using With Statement

```python
with open("Helloworld.txt", "r") as file:
    content = file.read()
    print(content)
```

Handling Exceptions When Closing a File

```python
try:
    file=open("Helloworld.txt","r")
    content=file.read()
    print(content)
except FileNotFoundError:
    print("File not found...")
finally:
    print("Closing a flie...")
```

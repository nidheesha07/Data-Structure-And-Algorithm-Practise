Insertion at Head

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
new_node = Node(5)
new_node.next = head
head = new_node
temp = head

while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

Insertion at Tail

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
new_node = Node(40)
temp = head

while temp.next:
    temp = temp.next

temp.next = new_node
temp = head

while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

Insertion in Middle

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
new_node = Node(25)
temp = head

while temp.data != 20:
    temp = temp.next

new_node.next = temp.next
temp.next = new_node
temp = head

while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create linked list
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

# Reverse the linked list
prev = None
curr = head

while curr:
    next_node = curr.next
    curr.next = prev
    prev = curr
    curr = next_node

head = prev

# Print reversed list
temp = head

while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

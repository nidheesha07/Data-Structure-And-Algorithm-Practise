```python
from collections import deque

graph = {
    1: [2, 3],
    2: [4],
    3: [],
    4: []
}

visited = []
queue = deque()

start = 1

visited.append(start)
queue.append(start)

while queue:
    node = queue.popleft()
    print(node, end=" ")

    for i in graph[node]:
        if i not in visited:
            visited.append(i)
            queue.append(i)
```

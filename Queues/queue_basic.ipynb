1. Queue using List

```python
queue = []

# Enqueue (Insert)
queue.append(10)
queue.append(20)
queue.append(30)

print("Queue after insertion:")
print(queue)

# Dequeue (Delete)
removed = queue.pop(0)

print("Removed element:", removed)
print("Queue after deletion:")
print(queue)

# Front element
print("Front element:", queue[0])

# Rear element
print("Rear element:", queue[-1])
```

Check if Queue is Empty

```python
queue = []

if len(queue) == 0:
    print("Queue is Empty")
else:
    print("Queue is Not Empty")
```

```python
queue = []

# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)

print("Queue:", queue)

# Front
print("Front:", queue[0])

# Rear
print("Rear:", queue[-1])

# Dequeue
print("Deleted:", queue.pop(0))

print("Queue after deletion:", queue)
```

```python
stack = []

# Push
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)

# Pop
print("Deleted:", stack.pop())
print(stack)
```

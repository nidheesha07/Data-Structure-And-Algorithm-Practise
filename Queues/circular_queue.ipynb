```python
queue = [10, 20, 30, 40, 50]
size = 5

front = 0
rear = 4

print("Initial Queue:", queue)

# Dequeue 10
front = (front + 1) % size

# Dequeue 20
front = (front + 1) % size

# Enqueue 60
rear = (rear + 1) % size
queue[rear] = 60

# Enqueue 70
rear = (rear + 1) % size
queue[rear] = 70

print("Queue after Circular Queue Operations:", queue)

print("Elements in Queue:", end=" ")
i = front
while True:
    print(queue[i], end=" ")
    if i == rear:
        break
    i = (i + 1) % size
```

```python
size = 5
queue = [None] * size

front = -1
rear = -1

# Enqueue
for item in [10, 20, 30, 40, 50]:
    if front == -1:
        front = rear = 0
    else:
        rear = (rear + 1) % size
    queue[rear] = item

print("Initial Queue:", queue)

# Dequeue two elements
front = (front + 1) % size
front = (front + 1) % size

# Enqueue 60 and 70
rear = (rear + 1) % size
queue[rear] = 60

rear = (rear + 1) % size
queue[rear] = 70

print("After Circular Queue:", queue)
```

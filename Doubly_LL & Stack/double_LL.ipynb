```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


# Create nodes
first = Node(10)
second = Node(20)
third = Node(30)

# Connect nodes
first.next = second
second.prev = first

second.next = third
third.prev = second

# Forward Traversal
print("Forward:")
temp = first
while temp:
    print(temp.data, end=" ")
    temp = temp.next

# Backward Traversal
print("\nBackward:")
temp = third
while temp:
    print(temp.data, end=" ")
    temp = temp.prev
```

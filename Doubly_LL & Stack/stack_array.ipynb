1. Push Operation

```python
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print(stack)
```

2. Pop Operation

```python
stack = [10, 20, 30]

item = stack.pop()

print("Removed:", item)
print("Stack:", stack)
```

3. Peek (Top Element)

```python
stack = [10, 20, 30]

print("Top element:", stack[-1])
```

4. Check if Stack is Empty

```python
stack = []

if len(stack) == 0:
    print("Stack is Empty")
else:
    print("Stack is Not Empty")
```

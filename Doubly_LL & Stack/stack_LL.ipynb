1. Push operation

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

top = None

def push(data):
    global top

    new_node = Node(data)
    new_node.next = top
    top = new_node

push(10)
push(20)
push(30)

temp = top

while temp:
    print(temp.data)
    temp = temp.next
```

```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Stack class
class Stack:
    def __init__(self):
        self.top = None

    # Push
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node
        print(data, "inserted")

    # Pop
    def pop(self):
        if self.top is None:
            print("Stack Underflow")
            return

        item = self.top.data
        self.top = self.top.next
        print(item, "removed")

    # Peek
    def peek(self):
        if self.top is None:
            print("Stack is Empty")
        else:
            print("Top element:", self.top.data)

    # Display
    def display(self):
        if self.top is None:
            print("Stack is Empty")
            return

        temp = self.top

        print("Stack elements:")
        while temp:
            print(temp.data)
            temp = temp.next


# Main
s = Stack()

s.push(10)
s.push(20)
s.push(30)

s.display()

s.peek()

s.pop()

s.display()
```

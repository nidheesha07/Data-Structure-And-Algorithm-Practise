```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

arr = [5, 2, 4, 1]
print(bubble_sort(arr))
```

Sort Ascending Order

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

nums = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(nums))
```

Sort Descending Order

```python
def bubble_sort_desc(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

nums = [5, 2, 8, 1, 9]
print(bubble_sort_desc(nums))
```

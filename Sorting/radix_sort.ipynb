```python
def radix_sort(arr):
    max_num = max(arr)
    exp = 1

    while max_num // exp > 0:
        buckets = [[] for _ in range(10)]

        for num in arr:
            digit = (num // exp) % 10
            buckets[digit].append(num)

        arr = []
        for bucket in buckets:
            arr.extend(bucket)

        exp *= 10

    return arr


numbers = [29, 83, 15, 47, 60]
print(radix_sort(numbers))
num = [129, 83, 315, 47,760 , 9]
print(radix_sort(num))
```

```python
arr = [21, 13, 45, 32]

# Sort by ones digit
arr = sorted(arr, key=lambda x: x % 10)

# Sort by tens digit
arr = sorted(arr, key=lambda x: (x // 10) % 10)

print(arr)
```

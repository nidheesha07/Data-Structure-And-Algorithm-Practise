Accessing an Array Element

```python
arr = [10, 20, 30, 40, 50]
print(arr[2])
```

Getting Length of a List

```python
numbers = [1, 2, 3, 4, 5]
print(len(numbers))
```

Updating an Element

```python
arr = [10, 20, 30, 40]
arr[1] = 100
print(arr)
```

Dictionary Lookup

```python
student={"name":"Nidhi" , "age":19}
print(student["name"])
```

Stack Peek Operation

```python
stack = [10, 20, 30, 40]
print(stack[-1])
print(stack[1])
```

using for loop

```python
# The loop always runs 5 times, regardless of input size.
for i in range(5):
    print(i)
```

Push Operation

```python
stack = [10, 20, 30]
stack.append(40)
print(stack)
```

Pop Operation

```python
stack = [10, 20, 30, 40]
stack.pop()
print(stack)
```

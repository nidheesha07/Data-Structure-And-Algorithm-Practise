Kadane's Algorithm (Maximum Subarray Sum)

```python
def kadane(arr):
    current_sum = arr[0]
    max_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
result = kadane(arr)
print(result)

arr1 = [1, 2, 3]
arr2 = [-2, 4, -1, 5]
print(kadane(arr1))
print(kadane(arr2))
```

Dutch National Flag Algorithm

```python
def dutch_national_flag(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1

        elif arr[mid] == 1:
            mid += 1

        else:  # arr[mid] == 2
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr


arr = [2, 0, 2, 3, 1, 1, 3, 0 , 1]

result = dutch_national_flag(arr)

print("Sorted Array:", result)
```

Missing numbers

```python
def find_missing_number(arr, n):
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(arr)

    return expected_sum - actual_sum


arr = [1, 2, 4, 5 ]
n = 5

print("Missing Number:", find_missing_number(arr, n))
```

Left Rotation

```python
def left_rotate(arr, k):
    k = k % len(arr)
    return arr[k:] + arr[:k]

arr = [1, 2, 3, 4, 5]

print(left_rotate(arr, 2))
```

Right Rotation

```python
def rotate_array(arr, k):
    k = k % len(arr)  # Handle k greater than array length
    return arr[-k:] + arr[:-k]

arr = [1, 2, 3, 4, 5]
k = 2

result = rotate_array(arr, k)

print("Rotated Array:", result)
```

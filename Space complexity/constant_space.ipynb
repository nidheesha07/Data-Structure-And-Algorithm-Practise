Find Maximum Element

```python
def find_max(arr):
    max_num = arr[0]

    for num in arr:
        if num > max_num:
            max_num = num

    return max_num

nums = [10, 5, 20, 8]
print(find_max(nums))
```

Find Minimum Element

```python
def find_min(arr):
    minimum = arr[0]

    for num in arr:
        if num < minimum:
            minimum = num

    return minimum
nums = [10, 5, 20, 8]
print(find_min(nums))
```

Count Even Numbers

```python
def count_even(arr):
    count = 0

    for num in arr:
        if num % 2 == 0:
            count += 1

    return count
```

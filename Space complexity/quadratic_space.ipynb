Create an n × n Matrix

```python
def create_matrix(n):

    matrix = []

    for i in range(n):
        row = [0] * n
        matrix.append(row)

    return matrix

print(create_matrix(3))
```

Multiplication Table Matrix

```python
def multiplication_table(n):

    table = []

    for i in range(1, n + 1):

        row = []

        for j in range(1, n + 1):
            row.append(i * j)

        table.append(row)

    return table

print(multiplication_table(4))
```

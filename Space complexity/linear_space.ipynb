1. Copy an Array

```python
def copy_array(arr):
    new_arr = []

    for num in arr:
        new_arr.append(num)

    return new_arr

nums = [1, 2, 3, 4, 5]
print(copy_array(nums))
```

2. Store Squares of Elements

```python
def squares(arr):
    result = []

    for num in arr:
        result.append(num * num)

    return result

nums = [1, 2, 3, 4]
print(squares(nums))
```

Reverse Array Using New List

```python
def reverse_array(arr):
    reversed_arr = []

    for i in range(len(arr)-1, -1, -1):
        reversed_arr.append(arr[i])

    return reversed_arr
arr=[10 , 20 , 30 , 40 , 50]
print(reverse_array(arr))
```

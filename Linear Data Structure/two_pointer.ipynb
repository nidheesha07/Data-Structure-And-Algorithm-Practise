Two Pointer Technique

```python
s="madam"
if s==s[::-1]:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
arr = [1, 2, 3, 4, 6, 8]
target = 10

left = 0
right = len(arr) - 1

while left < right:
    current_sum = arr[left] + arr[right]

    if current_sum == target:
        print("Pair found:", arr[left], arr[right])
        break

    elif current_sum < target:
        left += 1

    else:
        right -= 1
```

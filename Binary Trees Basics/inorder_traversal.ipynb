```python
class Node:

    def __init__(self, value):
        self.data = value
        self.left = None
        self.right = None


def inorder(root):
    if root:
        inorder(root.left)      # Visit left subtree
        print(root.data, end=" ")  # Visit root
        inorder(root.right)     # Visit right subtree

root = Node(10)

root.left = Node(20)
root.right = Node(30)

root.left.left = Node(40)
root.left.right = Node(50)

print("Inorder Traversal:")
inorder(root)
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return 0

numbers = [10, 25, 30, 45, 50]

target = 45

result = linear_search(numbers, target)

if result != 0:
    print("Element found at index", result)
else:
    print("Element not found")
```

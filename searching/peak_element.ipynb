Linear Search (Simple)

```python
arr = [1, 3, 20, 4, 1, 0]
n = len(arr)

for i in range(n):

    if i == 0:
        if arr[i] >= arr[i + 1]:
            print("Peak element:", arr[i])

    elif i == n - 1:
        if arr[i] >= arr[i - 1]:
            print("Peak element:", arr[i])

    else:
        if arr[i] >= arr[i - 1] and arr[i] >= arr[i + 1]:
            print("Peak element:", arr[i])
```

Binary Search (Optimal)

```python
arr = [1, 3, 20, 4, 1, 0]

low = 0
high = len(arr) - 1

while low < high:

    mid = (low + high) // 2

    if arr[mid] < arr[mid + 1]:
        low = mid + 1
    else:
        high = mid

print("Peak element:", arr[low])
```

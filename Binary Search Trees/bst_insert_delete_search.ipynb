```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, key):
    if root is None:
        return Node(key)
    if key < root.data:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root

def search(root, key):
    if root is None:
        return False
    if root.data == key:
        return True
    if key < root.data:
        return search(root.left, key)
    else:
        return search(root.right, key)

def delete(root, key):
    if root is None:
        return None
    if key < root.data:
        root.left = delete(root.left, key)
    elif key > root.data:
        root.right = delete(root.right, key)
    else:
        if root.left is None and root.right is None:
            return None
        else:
            print("Node has children. Simple delete can't delete it.")
    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

root = None
arr = [50, 30, 70, 20, 40, 60, 80]
for i in arr:
    root = insert(root, i)
print("BST:")
inorder(root)
print("\n")
if search(root, 60):
    print("60 Found")
else:
    print("60 Not Found")
root = delete(root, 20)
print("\nBST after deleting 20:")
inorder(root)
```

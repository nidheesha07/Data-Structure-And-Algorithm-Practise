```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def validate(root, minimum, maximum):
    if root is None:
        return True

    if root.data <= minimum or root.data >= maximum:
        return False

    return (validate(root.left, minimum, root.data) and
            validate(root.right, root.data, maximum))


# BST
root = Node(50)
root.left = Node(30)
root.right = Node(70)
root.left.left = Node(20)
root.left.right = Node(40)
root.right.left = Node(60)
root.right.right = Node(80)

if validate(root, float('-inf'), float('inf')):
    print("Valid BST")
else:
    print("Not a BST")
```
